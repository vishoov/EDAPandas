# One-Hot Encoding


```python
category_columns = df1.select_dtypes(include = ['object']).columns
category_columns 
```

```python
df1 = pd.get_dummies(df1, columns = category_columns, drop_first=True)
```

```python
df1.shape
```

```python
df1.head()
```

## Split the data into dependent and independent variable

```python
x = df1.drop(['Yield'], axis = 1)
y = df1[['Yield']]
```

```python
print(x.shape)
y.shape
```

```python
x.head()
```

```python
y.head()
```

## Splitting the data set into train and test set

```python
#split the data into training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape, x_test.shape, y_train.shape,y_test.shape
```

# Power Transformation using the method 'Yeo-Johnson'

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')

x_train_transform1 = pt.fit_transform(x_train)
x_test_transform1 = pt.fit_transform(x_test)
```

```python
df_trans = pd.DataFrame(x_train_transform1, columns=x_train.columns)
df_trans.head()
```

## After Transformation, there is no need for Standardization of the data

```python
plt.figure(figsize=(15,20))
plt.subplot(4,2,1)
sns.distplot(df_trans['Area'],bins = 20,color = 'red')
plt.subplot(4,2,2)
sns.distplot(df_trans['Production'],bins = 10,color = 'green')
plt.subplot(4,2,3)
sns.distplot(df_trans['Annual_Rainfall'],bins = 10,color = 'fuchsia')
plt.subplot(4,2,4)
sns.distplot(df_trans['Fertilizer'],bins = 10, color = 'indigo')
```

## Viewing the Q-Q Plot after the Transformation

```python
plt.figure(figsize=(15,20))
plt.subplot(4,2,1)
stats.probplot(df_trans['Area'], dist = 'norm', plot = plt)
plt.subplot(4,2,2)
stats.probplot(df_trans['Production'], dist = 'norm', plot = plt)
plt.subplot(4,2,3)
stats.probplot(df_trans['Annual_Rainfall'], dist = 'norm', plot = plt)
plt.subplot(4,2,4)
stats.probplot(df_trans['Fertilizer'], dist = 'norm', plot = plt)

plt.show()
```

# Linear Regression with skewed data

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred_train = lr.predict(x_train)
print("Training Accuracy : ",r2_score(y_train,y_pred_train))

y_pred_test = lr.predict(x_test)
print("Test Accuracy : ",r2_score(y_test,y_pred_test))
```

# Time Series Analysis

### We will use ARIMA (AutoRegressive Integrated Moving Average) for this purpose. 
### Since ARIMA requires time series data, let's convert the 'Crop_Year' column to datetime and set it as the index:

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

df = df.sort_index()
df.index = pd.date_range(start=df.index.min(), periods=len(df), freq='D')

# Visualize the time series
plt.figure(figsize=(12, 6))
plt.plot(df['Yield'])
plt.title('Yield Over Time')
plt.xlabel('Date')
plt.ylabel('Yield')
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming df is your DataFrame with the 'Yield' column
# Make sure the 'Date' column is in datetime format and set it as the index
# Example: df['Date'] = pd.to_datetime(df['Date'])
# df.set_index('Date', inplace=True)

# Ensure the index is monotonic and has a frequency (assuming daily frequency)
df = df.sort_index()
df.index = pd.date_range(start=df.index.min(), end=df.index.max(), freq='D')

# Decompose the time series
result = seasonal_decompose(df['Yield'], model='additive', period=1)

# Access the decomposed components
trend = result.trend
seasonal = result.seasonal
residual = result.resid

# Create a new DataFrame to store the decomposed components
decomposed_df = pd.DataFrame({
    'Date': df.index,
    'Yield': df['Yield'],
    'Trend': trend,
    'Seasonal': seasonal,
    'Residual': residual
})

# Save the decomposed data to a CSV file or any desired format
decomposed_df.to_csv('decomposed_yield_data.csv', index=False)
```

```python
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Assuming df is your DataFrame with the 'Yield' column
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df_numeric = df[numeric_columns]

# Replace missing values with forward fill
df_numeric = df_numeric.ffill()

# Sort the DataFrame by the date index in ascending order
df_numeric.sort_index(inplace=True)

# Assuming 'Yield' is the column you want to perform time series analysis on
ts_data = df_numeric['Yield']

# Create a new date index with the desired frequency
new_index = pd.date_range(start=df_numeric.index.min(), end=df_numeric.index.max(), freq='D')

# Perform time series analysis (Exponential Smoothing)
model = ExponentialSmoothing(ts_data, seasonal='add', seasonal_periods=365)  
# Adjust parameters as needed
fitted_model = model.fit()

# Forecast future values
forecast_start = new_index.max() + pd.Timedelta(days=1)
forecast_end = forecast_start + pd.DateOffset(days=364)  # Adjust the number of days as needed
future_index = pd.date_range(start=forecast_start, end=forecast_end, freq='D')
future_values = fitted_model.forecast(steps=len(future_index))

# Create a DataFrame with the forecasted values
forecast_df = pd.DataFrame({'Forecasted_Yield': future_values}, index=future_index)
# Save the forecasted data to an Excel file
forecast_df.to_excel('forecasted_yield_data.xlsx')
```
